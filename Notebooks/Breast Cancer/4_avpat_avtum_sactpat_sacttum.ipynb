```python
import pandas as pd
import numpy as np
from zipfile import ZipFile
```

```python
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
zf = ZipFile("../../Data/sim_sact_tumour.zip")
csv = zf.open("sim_sact_tumour.csv")
sim_sact_tumour = pd.read_csv(csv)
sim_sact_tumour.head()
```

```python
sim_sact_tumour.shape
```

```python
print("The original dataset has data for %d patients" % len(sim_sact_tumour))
print("After removing duplicates, the dataset has data for %d patients" % len(sim_sact_tumour.drop_duplicates()))
print("There are no duplicates in the dataset")
```

```python
sim_sact_tumour.columns
```

```python
columns_selected = ['MERGED_TUMOUR_ID', 'MERGED_PATIENT_ID']
sim_sact_tumour = sim_sact_tumour[columns_selected]
sim_sact_tumour.head()
```

```python
print("The original dataset has data for %d patients" % len(sim_sact_tumour))
print("After removing duplicates, the dataset has data for %d patients" % len(sim_sact_tumour.drop_duplicates()))
print("There are no duplicates in the dataset")
```

```python
df = pd.read_csv("./Data/avpat_avtum_sactpat.csv", sep = '^', index_col = 0)
df.head()
```

```python
df.shape
```

```python
df = df.merge(sim_sact_tumour, left_on = 'MERGED_PATIENT_ID', right_on = 'MERGED_PATIENT_ID', how = 'left')
df.head()
```

```python
df = df.drop_duplicates()
df.shape
```
