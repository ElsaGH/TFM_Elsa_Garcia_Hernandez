```python
def behaviour(value):
    if value.startswith('0'):
        return 'Benign'
    elif value.startswith('1'):
        return 'Uncertain'
    elif value.startswith('2'):
        return 'In situ'
    elif value.startswith('3'):
        return 'Malignant'
    elif value.startswith('5'):
        return 'Micro-invasive'
    elif value.startswith('6'):
        return 'Malignant, metastatic/secondary site'
    elif value.startswith('9'):
        return 'Malignant, uncertain whether primary or metastatic'
    elif value.startswith('X'):
        return 'Unknown/inapplicable'
    else:
        return value

df['BEHAVIOUR_ICD10_O2'] = df['BEHAVIOUR_ICD10_O2'].apply(behaviour)
        
df['BEHAVIOUR_ICD10_O2'].value_counts()
```

```python
df['BEHAVIOUR_ICD10_O2'] = df['BEHAVIOUR_ICD10_O2'].astype(str)
```

```python
#Since I am selecting patients with only 1 tumour, having a metastatic tumour is not possible
df['BEHAVIOUR_ICD10_O2'].replace(['Malignant, metastatic/secondary site','Malignant, uncertain whether primary or metastatic'], np.nan, inplace = True)
```

```python
df['BEHAVIOUR_ICD10_O2'].replace('Unknown/inapplicable', np.nan, inplace = True)
```

```python
df['T_BEST'] = df['T_BEST'].astype(str)
```

```python
def T_best(value):
    if value == 'T2':
        return '2'
    elif value == '2A':
        return '2a'
    elif value == '1B':
        return '1b'
    elif value == 'T1':
        return '1'
    elif value == '1C':
        return '1c'
    elif (value == 'X' or value == 'IS' or value == 'is' or value == 'a'):
        return np.nan
    else:
        return value

df['T_BEST'] = df['T_BEST'].apply(T_best)
```

```python
df['N_BEST'] = df['N_BEST'].astype(str)
```

```python
def N_best(value):
    if value == '0.0':
        return '0'
    elif value == '1.0':
        return '1'
    elif value == 'N0':
        return '0'
    elif value == '2.0':
        return '2'
    elif value == '3.0':
        return '3'
    elif value == '00':
        return '0'
    elif value == 'X':
        return np.nan
    else:
        return value

df['N_BEST'] = df['N_BEST'].apply(N_best)
```
